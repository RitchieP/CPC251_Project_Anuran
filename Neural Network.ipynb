### Feature Selection

We used a heatmap to help us visualize the correleration of each feature to it's output. As we can see from the heatmap generated below, almost every feature have a correleration to each other.

After experimenting with feature selection, we found out that by using all the features available produces a better result than selecting less features for the neural network. Hence, our code will be using all the features available.

````python
import seaborn as sns

sns.heatmap(dataset.corr())
````

````python
X = dataset.iloc[:, 0:22]
y = dataset.iloc[:, 22]
X.head(2)
````

### Split dataset

We will now spit the dataset into training and test set

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Further split training data set into validation data set
X_test, X_valid, y_test, y_valid = train_test_split(X_test, y_test, test_size=0.5)

# Print to check data set size
print(X_train.shape)
print(X_test.shape)
print(X_valid.shape)
````

### Training model and predicting

For this section, we will be using a Multi-Layer Perceptron function provided by the sklearn library.

We changed the maximum iterations for the model to 500 because if the number is too low, there will be a optimization haven't converge warning.

````python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(max_iter=500)
# Training the model
model.fit(X_train, y_train)
# Predicting with the model
y_pred = model.predict(X_valid)
````

### Model Evaluation

This section we will evaluate the model

````python
# Function to check accuracy of algorithm

def accuracy(confusion_matrix):
   diagonal_sum = confusion_matrix.trace()
   sum_of_all_elements = confusion_matrix.sum()
   return diagonal_sum / sum_of_all_elements
````

````python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_pred, y_valid))
print(confusion_matrix(y_pred, y_valid))

print("Accuracy of MLPClassifier : ", accuracy(confusion_matrix(y_pred, y_valid)))
````
